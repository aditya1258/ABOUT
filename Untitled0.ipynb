LOADING THE DATASET

This dataset contains information about used cars. This data can be used for a lot of purposes such as price prediction to exemplify the use of linear regression in Machine Learning. The columns in the given dataset are as follows:

name

year

selling_price

km_driven

fuel

seller_type

transmission

Owner

```python
import pandas as pd

dataset = pd.read_csv(
    'https://raw.githubusercontent.com/Kraghav2002/Machine-Learning---Core-Concepts/main/CAR%20DETAILS%20FROM%20CAR%20DEKHO.csv', 
    # dataset has an extra index columns. We don't need it. 
    # Just load selling_price and km_driven
    usecols=[2, 3] 
)
dataset.info()

```

```python
import seaborn as sns 

sns.scatterplot(
    x=dataset['selling_price'], 
    y=dataset['km_driven'],
)
```

BUILDING REGRESSION MODEL

```python
# Input
X = dataset['selling_price']
# Output
y = dataset['km_driven']
```

SPLITTING THE DATASET TO TRAIN AND TEST DATA

```python
from sklearn.model_selection import train_test_split

# 67% - training set (X_train, y_train)
# 33% - test set (X_test, y_test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)

# X_train and X_test are instances of pandas Series because
# they contain only one column. Convert them to DataFrames
X_train = X_train.to_frame()
X_test = X_test.to_frame()
```

CREATING AND TRAINING THE DATASET

```python
from sklearn.linear_model import LinearRegression
# Create a new model
model = LinearRegression()
# build the model using the traing data
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)

actual_vs_predicted = pd.DataFrame(
    {'Actual': y_test, 
     'Predicted':predicted}
)
# Show first 5 rows
actual_vs_predicted.head().round(2)
```

```python
# Make a prediction for a single input
sample_input = [250]
prediction = model.predict([sample_input])

# Print the prediction
print('The predicted price for the input is:', prediction[0])
```

MAE (MEAN ABSOLUTE ERROR BEFORE NORMALIZATION)

```python
from sklearn.metrics import (
    mean_absolute_error, # MAE
    mean_squared_error # MSE
)
```

```python
MAE = mean_absolute_error(
    y_true=y_test, # actual values
    y_pred=predicted # predicted values
)
MAE.round(2)
```

MSE (MEAN SQUARED ERROR BEFORE NORMALIZATION)

```python
MSE = mean_squared_error(
    y_true=y_test, # actual values
    y_pred=predicted # predicted values
)
MSE.round(2)
```

RMSE (ROOT MEAN SQUARED ERROR BEFORE NORMALIZATION)

```python
RMSE = MSE**(1/2)
RMSE.round(2)
```

NORMALIZATION USING MIN_MAX_SCALER


```python
from sklearn.preprocessing import MinMaxScaler

# Extract the numerical features to be normalized
numerical_features = ['selling_price', 'km_driven']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
dataset[numerical_features] = scaler.fit_transform(dataset[numerical_features])

# Print the normalized data
print(dataset)
```

MAE (MEAN ABSOLUTE ERROR AFTER NORMALIZATION)

```python
MAE = mean_absolute_error(
    y_true=y_test, # actual values
    y_pred=predicted # predicted values
)
MAE.round(2)
```

MSE (MEAN SQUARED ERROR AFTER NORMALIZATION)

```python
MSE = mean_squared_error(
    y_true=y_test, # actual values
    y_pred=predicted # predicted values
)
MSE.round(2)
```

MSE (ROOT MEAN SQUARED ERROR AFTER NORMALIZATION)

```python
RMSE = MSE**(1/2)
RMSE.round(2)
```
